```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None, names=["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm", "class"])
df.insert(0, "id", range(1, len(df) + 1))
df
```

```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
df['class'] = lbl.fit_transform(df['class'])
df.head(100)
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
df.plot(kind='scatter', x='SepalLengthCm', y='PetalLengthCm', c='teal', s=30)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

class_colors = {
    'Iris-setosa': 'blue',
    'Iris-versicolor': 'orange',
    'Iris-virginica': 'green'
}
g = sns.FacetGrid(df, hue="class", height=6)
g.map(plt.scatter, 'SepalLengthCm', 'PetalLengthCm')

legend_labels = []
for class_name, color in class_colors.items():
    plt.scatter([], [], label=class_name, color=color)
    legend_labels.append(class_name)
    
g.add_legend(labels=legend_labels, title="Classes")

plt.show()
```

```python
x = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].values
y = df[["class"]].values
```

```python
#TRAINING
```

```python
from sklearn.model_selection import train_test_split as tts

x_train, x_test, y_train, y_test = tts(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
#TESTING
```

```python
x_predict = lr.predict(x_test)
x_predict
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn import metrics
Actual = y
predicted = lr.predict(x)
print(metrics.classification_report(Actual, predicted))
```
